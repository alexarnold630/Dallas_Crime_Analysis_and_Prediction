# Clean the Data

```python
# split lat and long
lats = []
longs = []

for _, row in df.iterrows():
    try:
        clean_row = row["geocoded_column"].replace("'", '"').replace('"{', "{").replace('}"', "}")
        clean_row = json.loads(clean_row)
        latitude = clean_row['latitude']
        lats.append(latitude)
        longitude = clean_row['longitude']
        longs.append(longitude)
    except:
        lats.append("")
        longs.append("")

df["latitude"] = lats
df["longitude"] = longs
df.drop(["geocoded_column"], axis=1)
df.head()
```

```python
# change Dtype

for col in df:
    df[col] = pd.to_numeric(df[col], errors='ignore')
    
df["date1"] = pd.to_datetime(df["date1"])
df["date2_of_occurrence_2"] = pd.to_datetime(df["date2_of_occurrence_2"])
df["reporteddate"] = pd.to_datetime(df["reporteddate"])
df["edate"] = pd.to_datetime(df["edate"] + " " + df["etime"])
df["callorgdate"] = pd.to_datetime(df["callorgdate"])
df["callreceived"] = pd.to_datetime(df["callreceived"])
df["callcleared"] = pd.to_datetime(df["callcleared"])
df["calldispatched"] = pd.to_datetime(df["calldispatched"])
df["time1"] = pd.to_datetime(df["time1"])
df["time2"] = pd.to_datetime(df["time2"])
df["etime"] = pd.to_datetime(df["etime"])
df.info()
```

```python
#Clean Columns
# clean status
df2 = df[(df.status != 'Open') & (df.status != 'L') & (df.status !="Returned for Correction")]
df2 = df2.replace("Clear by Exceptional Arrest", "Clear by Arrest")
df2.dropna(subset=['status'], how="all", inplace=True)

# clean division
df2 = df2.replace("CENTRAL", "Central")
df2 = df2.replace("NORTH CENTRAL", "NorthCentral")
df2 = df2.replace("NORTHEAST", "NorthEast")
df2 = df2.replace("NORTHWEST", "NorthWest")
df2 = df2.replace("SOUTH CENTRAL", "SouthCentral")
df2 = df2.replace("SOUTHEAST", "SouthEast")
df2 = df2.replace("SOUTHWEST", "SouthWest")
df2.dropna(subset=['division'], how="all", inplace=True)

#clean district
df2.dropna(subset=['district'], how="all", inplace=True)
df2 = df2.replace("9", "D9")
df2 = df2.replace("10", "D10")
df2 = df2.replace("11", "D11")

#clean zip code
df2.dropna(subset=['zip_code'], how="all", inplace=True)
line = df2.loc[df2.zip_code== 12345, :].index
df2.drop(line, inplace=True)

#split time to hours
df2['hour'] = df2["time1"].dt.hour

df2.head()
```

```python
#drop zip codes
df2 = df2.drop(df2[df2.zip_code.isin(["75007", "75006", "75080", "75042", "76203", "75019", "75181", "75041", "75051", "75116", "75115", "75134", "75001", "75062", "75088", "75159", "75043", "75150", "75081", "75075", "75180", "75127", "75063", "75409", "33436", "75052", "75406", "75056", "76182", "75137", "76226", "75104", "73237", "75149", "76034"])].index)
```
